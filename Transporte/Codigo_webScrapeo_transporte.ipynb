```python
# Importación de librerías necesarias
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.edge.service import Service as EdgeService
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.microsoft import EdgeChromiumDriverManager
import pandas as pd
import time

# Lista de departamentos a consultar (destinos)
destinos = [
    "Amazonas", "Ancash", "Apurímac", "Arequipa", "Ayacucho", "Cajamarca", "Callao", "Cusco",
    "Huancavelica", "Huánuco", "Ica", "Junín", "La Libertad", "Lambayeque", "Loreto", "Madre de Dios",
    "Moquegua", "Pasco", "Piura", "Puno", "San Martín", "Tacna", "Tumbes"
]
# Lista vacía donde se almacenarán los resultados
datos = []

# Inicializar el navegador Microsoft Edge con WebDriver Manager (no necesitas descargar el driver manualmente)
driver = webdriver.Edge(service=EdgeService(EdgeChromiumDriverManager().install()))
driver.maximize_window()# Maximizar ventana para asegurar carga completa de la página

try:# Iterar sobre cada destino
    for destino in destinos:
        origen = "Lima"# Punto de partida
        print(f"Consultando {origen} → {destino} ...")
        url = f"https://www.rome2rio.com/map/{origen}/{destino}"# Armar la URL de consulta en Rome2Rio
        driver.get(url)

        # Esperar hasta que se carguen los bloques de transporte (máximo 20 segundos)
        WebDriverWait(driver, 20).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, "div.sc-dGlnUf"))
        )
        time.sleep(4)# Esperar hasta que se carguen los bloques de transporte (máximo 20 segundos)

        # Scroll para cargar todo
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(4)
        # Encontrar todos los bloques de transporte disponibles en la página
        bloques = driver.find_elements(By.CSS_SELECTOR, "div.sc-dGlnUf")

        # Recorrer cada bloque y extraer los datos necesarios
        for bloque in bloques:
            # Transporte
            try:
                transporte = bloque.find_element(By.CSS_SELECTOR, "svg title").get_attribute("textContent")
            except:
                transporte = "Desconocido"

            # Duración estimada del viaje
            try:
                duracion = bloque.find_element(By.CSS_SELECTOR, "time").text
            except:
                duracion = "No disponible"
            # Guardar los datos obtenidos en un diccionario y agregar a la lista
            datos.append({
                "Origen": origen,
                "Destino": destino,
                "Transporte": transporte,
                "Duración Estimada": duracion
            })

finally:
    driver.quit()# Cerrar el navegador al finalizar el scraping (tanto si ocurre un error como si todo sale bien)

# Crear DataFrame
df = pd.DataFrame(datos)

# Mostrar tabla
pd.set_option("display.max_rows", None)
display(df)


# Guardar en Excel
df.to_excel("viajes_lima_destinos10.xlsx", index=False)
```
